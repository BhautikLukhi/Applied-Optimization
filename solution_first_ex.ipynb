```python
#1a
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

delta = 0.005
x = y = np.arange(-10.0, 10.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = (X**2 + Y**2)
Z2 = np.cos(3*X - 3*Y)
Z3 = np.cos(3*X + 3*Y)
Z = 1/100*Z1 - 1/2*(Z2 + Z3)


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.pcolormesh(X, Y, Z, shading='auto', cmap='viridis')
plt.colorbar(label='f(x, y)')
plt.title('Heat Map of f(x, y)')
plt.xlabel('x')
plt.ylabel('y')


plt.tight_layout()
plt.show()
```

1b
it is a bit hard to describe the exact shape of the first function due to the wide variety of the values of f(x,y), there are a lot of jumps with more higher values on the borders and lower ones on the center, there is no clear one global optima from the heat map, it has more than one local optimum

```python
from scipy.optimize import minimize
import random as rd
def funt(X):
    return 1/100*(X[0]**2 + X[1]**2) - 1/2*(np.cos(3*X[0] - 3*X[1]) + np.cos(3*X[0] + 3*X[1]))
x_min, x_max = -10, 10
y_min, y_max = -10, 10

initial_guess = [rd.uniform(x_min, x_max), rd.uniform(y_min, y_max)]
print("Random Initial Guess (x, y):", initial_guess)

result = minimize(funt, initial_guess, method='Nelder-Mead')

print("Optimization Result:")
print("Optimal Solution:", result.x)
print("Optimal Value:", result.fun)
```

```python
p0s = []
p1s =[]
for n in range(0,10):
  def funt(X):
    return 1/100*(X[0]**2 + X[1]**2) - 1/2*(np.cos(3*X[0] - 3*X[1]) + np.cos(3*X[0] + 3*X[1]))
  x_min, x_max = -10, 10
  y_min, y_max = -10, 10

  initial_guess = [rd.uniform(x_min, x_max), rd.uniform(y_min, y_max)]
  result = minimize(funt, initial_guess, method='Nelder-Mead')
  p0s.append(initial_guess)
  p1s.append((result.x).tolist())
print(p0s)
print(p1s)
```

```python
#c
from scipy.optimize import minimize
import random as rd
def funt(x):
    k, h, b = x
    return -2.5 * k * (1/27 * h * b**2) + 0.75 * k * (2 * b * h) + 0.2 * k * (2 * b)**2


initial_guess = [rd.uniform(10, 30), rd.uniform(10, 30), rd.uniform(10, 30)]
print("Random Initial Guess (x, y):", initial_guess)

result = minimize(funt, initial_guess, method='Nelder-Mead')

print("Optimization Result:")
print("Optimal Solution:", result.x)
print("Optimal Value:", result.fun)
```
